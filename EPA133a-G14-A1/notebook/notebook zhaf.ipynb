```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the TSV file

file_path = r"C:\Users\zhafr\Desktop\Q3\Advanced Simulation\Week 1-2 Assignments\WBSIM_Lab1_2024\WBSIM_Lab1_2024\infrastructure\_roads.tsv"

# Load the file using sep='\t' for Tab-Separated Values
df = pd.read_csv(file_path, sep='\t')
#df = pd.read_csv(r'C:\Users\zhafr\Desktop\ASIM\assignment 1\WBSIM_Lab1_2024\infrastructure\_roads.tsv', sep='\t')

# Display the dataframe
df
```

```python
# Unpivot the dataframe to create a tidier format
melted_data = []

for idx, row in df.iterrows():
    road = row['road']
    
    # Start from column 1 to include lrp1, iterate through columns in groups of 3
    for i in range(1, len(df.columns), 3):
        if i + 2 < len(df.columns):
            lrp = row.iloc[i]
            lat = row.iloc[i + 1]
            lon = row.iloc[i + 2]
            
            # Only add if coordinates are not zero
            if not ((lat == 0) and (lon == 0)):
                melted_data.append({'road': road, 'lrp': lrp, 'lat': lat, 'lon': lon})

# Create new tidy dataframe
df_tidy = pd.DataFrame(melted_data)
df_tidy
```

```python
n1_n2_coords = df_tidy[df_tidy['road'].str.contains('^N1$|^N2$', case=True, na=False)].dropna(subset=['lat', 'lon'])

reference_pool = set(zip(n1_n2_coords['lat'], n1_n2_coords['lon']))

roads_with_nas = df_tidy[df_tidy['lat'].isna() | df_tidy['lon'].isna()]['road'].unique()

intersect_list = []
no_intersect_list = []

# Iterate through each road that has missing data
for road_id in roads_with_nas:
    current_road_coords = df_tidy[df_tidy['road'] == road_id].dropna(subset=['lat', 'lon'])
    current_coords_set = set(zip(current_road_coords['lat'], current_road_coords['lon']))
    
    if not current_coords_set.isdisjoint(reference_pool):
        intersect_list.append(road_id)
    else:
        no_intersect_list.append(road_id)

# Output Results
print(f"Roads with NaNs that intersect N1/N2: {(intersect_list)}")
print(f"Roads with NaNs that are isolated: {(no_intersect_list)}")
```

To find which roads intersect with N1 and/or N2

```python

n1_n2_mask = df_tidy['road'].str.contains('^N1$|^N2$', case=True, na=False)
n1_n2_coords = df_tidy[n1_n2_mask]
reference_pool = set(zip(n1_n2_coords['lat'].round(4), n1_n2_coords['lon'].round(4)))


all_other_roads = df_tidy[~n1_n2_mask]['road'].unique()

intersecting_roads = []
isolated_roads = []


for road_id in all_other_roads:
    current_road = df_tidy[df_tidy['road'] == road_id]
    current_coords = set(zip(current_road['lat'].round(4), current_road['lon'].round(4)))
    
    if not current_coords.isdisjoint(reference_pool):
        intersecting_roads.append(road_id)

print(f"Roads intersecting N1 or N2: {intersecting_roads}")


```

```python
def detect_bridges_wrong_coordinates(df, lat_col='lat', lon_col='lon', road_col='road', lrp_col='LRPName'):

    # Define the bounding box for Bangladesh DOUBLE CHECK IF THESE ARE CORRECT!!!!!!!!
    min_lat, max_lat = 20.86382, 26.33338
    min_lon, max_lon = 88.15638, 92.30153

    # Create flags for inverted coordinates and far-out coordinates
    df['inverted_coords'] = (
        (df[lat_col] >= min_lon) & (df[lat_col] <= max_lon) &  # Latitude in longitude range
        (df[lon_col] >= min_lat) & (df[lon_col] <= max_lat)    # Longitude in latitude range
    )

    df['far_out_coords'] = ( #can we assume that the ones that are inverted are also far out? Maybe not, but we can check for both just in case
        ((df[lat_col] < min_lat) | (df[lat_col] > max_lat)) &  # Latitude outside range
        ((df[lon_col] < min_lon) | (df[lon_col] > max_lon))    # Longitude outside range
    )

    # Filter rows where coordinates are either inverted or far out
    outside_bounding_box = df[df['inverted_coords'] | df['far_out_coords']]

    # Get unique bridges/roads outside the bounding box
    
    #bridges_outside_bbox = outside_bounding_box[road_col].unique()
    bridges_outside_bbox = outside_bounding_box[[road_col, lrp_col]].drop_duplicates().values.tolist() #Need to get the LRPname as well to know which bridge it is

    return outside_bounding_box, bridges_outside_bbox
```

```python
# Call the function
outside_bbox_df, bridges_outside_bbox = detect_bridges_wrong_coordinates(sheet_df)

# Display the results
print("Bridges with latitudes and longitudes outside the bounding box of Bangladesh:")
print(bridges_outside_bbox)

# Optionally, display the full rows for inspection
display(outside_bbox_df)
```
