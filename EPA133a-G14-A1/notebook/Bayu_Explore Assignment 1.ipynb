```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the TSV file
df = pd.read_csv(r'C:\Users\bjama\Desktop\ASIM\assignment 1\WBSIM_Lab1_2024\infrastructure\_roads.tsv', sep='\t')

# Display the dataframe
df
```

```python
# Unpivot the dataframe to create a tidier format
melted_data = []

for idx, row in df.iterrows():
    road = row['road']
    
    # Start from column 1 to include lrp1, iterate through columns in groups of 3
    for i in range(1, len(df.columns), 3):
        if i + 2 < len(df.columns):
            lrp = row.iloc[i]
            lat = row.iloc[i + 1]
            lon = row.iloc[i + 2]
            
            # Only add if coordinates are not zero
            if not ((lat == 0) and (lon == 0)):
                melted_data.append({'road': road, 'lrp': lrp, 'lat': lat, 'lon': lon})

# Create new tidy dataframe
df_tidy = pd.DataFrame(melted_data)
df_tidy
```

```python
# Find duplicates with same road and lrp values in df_tidy, excluding NaN lrp values
duplicates = df_tidy[df_tidy.duplicated(subset=['road', 'lrp'], keep=False) & df_tidy['lrp'].notna()].sort_values(['road', 'lrp'])
duplicates
```

```python
# Filter df_tidy for road N1
n1_data = df_tidy[df_tidy['road'] == 'N1']

# Create a plot
plt.figure(figsize=(12, 8))
plt.plot(n1_data['lon'], n1_data['lat'], marker='o', linestyle='-', markersize=3)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Road N1 - Geographic Path')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from math import radians, cos, sin, asin, sqrt

def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    Returns distance in kilometers
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    
    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    km = 6371 * c
    return km

# Create a copy of n1_data
n1_data_cleaned = n1_data.copy()

# Calculate distances between consecutive points
distances = []
for i in range(len(n1_data_cleaned) - 1):
    dist = haversine(
        n1_data_cleaned.iloc[i]['lon'],
        n1_data_cleaned.iloc[i]['lat'],
        n1_data_cleaned.iloc[i+1]['lon'],
        n1_data_cleaned.iloc[i+1]['lat']
    )
    distances.append(dist)

# Calculate median and threshold for outliers
median_dist = np.median(distances)
threshold = median_dist * 10  # Points more than 10x the median distance are outliers

# Identify and interpolate outliers
for i in range(1, len(n1_data_cleaned) - 1):
    dist_prev = haversine(
        n1_data_cleaned.iloc[i-1]['lon'],
        n1_data_cleaned.iloc[i-1]['lat'],
        n1_data_cleaned.iloc[i]['lon'],
        n1_data_cleaned.iloc[i]['lat']
    )
    
    dist_next = haversine(
        n1_data_cleaned.iloc[i]['lon'],
        n1_data_cleaned.iloc[i]['lat'],
        n1_data_cleaned.iloc[i+1]['lon'],
        n1_data_cleaned.iloc[i+1]['lat']
    )
    
    # If both distances are large, interpolate
    if dist_prev > threshold or dist_next > threshold:
        # Linear interpolation between previous and next point
        n1_data_cleaned.iloc[i, n1_data_cleaned.columns.get_loc('lat')] = \
            (n1_data_cleaned.iloc[i-1]['lat'] + n1_data_cleaned.iloc[i+1]['lat']) / 2
        n1_data_cleaned.iloc[i, n1_data_cleaned.columns.get_loc('lon')] = \
            (n1_data_cleaned.iloc[i-1]['lon'] + n1_data_cleaned.iloc[i+1]['lon']) / 2

# Plot the cleaned data
plt.figure(figsize=(12, 8))
plt.plot(n1_data_cleaned['lon'], n1_data_cleaned['lat'], marker='o', linestyle='-', markersize=3, label='Cleaned')
plt.plot(n1_data['lon'], n1_data['lat'], marker='x', linestyle='--', markersize=2, alpha=0.5, label='Original')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Road N1 - Geographic Path (Outliers Interpolated)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Median distance between consecutive points: {median_dist:.2f} km")
print(f"Outlier threshold: {threshold:.2f} km")
```
